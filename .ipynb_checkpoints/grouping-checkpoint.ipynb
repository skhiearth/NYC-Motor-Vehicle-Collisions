```python
import numpy as np
import pandas as pd
import os
import calendar
import datetime
```

```python
# Importing static dataset from .csv file
df = pd.read_csv('datasets/Motor_Vehicle_Collisions_-_Crashes_Clean.csv')

print(df.shape)
df.head()
```

```python
df = df.sort_values(['CRASH_DATE', 'BOROUGH'])
print(df.shape)
df.head(30)
```

```python
df['VEHICLE TYPE CODE 1'].value_counts()[:5]
```

```python
df = df[df['VEHICLE TYPE CODE 1'].isin(['PASSENGER VEHICLE', 'Sedan', 'SPORT UTILITY / STATION WAGON', 'Station Wagon/Sport Utility Vehicle', 'TAXI'])]
```

```python
print(df.shape)
df.head()
```

```python
df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts()[:10]
```

```python
df = df[df['CONTRIBUTING FACTOR VEHICLE 1'].isin(['Unspecified', 'Driver Inattention/Distraction', 'Failure to Yield Right-of-Way', 'Backing Unsafely', 'Following Too Closely', 'Other Vehicular', 'Passing Too Closely', 'Passing or Lane Usage Improper', 'Turning Improperly', 'Unsafe Lane Changing'])]
```

```python
print(df.shape)
df.head()
```

```python
cols_to_drop = ['index']
df.drop(cols_to_drop, axis = 1, inplace = True)

df = df.reset_index()

cols_to_drop = ['index']
df.drop(cols_to_drop, axis = 1, inplace = True)

df.head()
```

```python
df["Accidents"] = 1
df = df.groupby(["CRASH_DATE", "BOROUGH", "VEHICLE TYPE CODE 1", "CONTRIBUTING FACTOR VEHICLE 1"])["Accidents"].count().reset_index()
```

```python
print(df.shape)
df.head(20)
```

```python
df.to_csv('datasets/Motor_Vehicle_Collisions_-_Crashes_Grouped.csv')
```
